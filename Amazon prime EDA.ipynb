Project Summary: Exploratory Data Analysis on Amazon prime Titles and Credits Dataset
This project focuses on performing Exploratory Data Analysis (EDA) Amazon prime’s content catalog using two key datasets: one containing title-level metadata and the other including information about the associated cast and crew. The main objective is to uncover trends, patterns, and insights from Amazon prime’s content library using Python and key data science libraries. These insights help us better understand the nature of the content, popular genres, IMDb scores, and involvement of directors and actors.

Datasets Overview
The two datasets used are:

titles.csv: Contains metadata about Amazon prime titles, including attributes like title name, type (Movie or TV Show), release year, genres, runtime, production country, and IMDb score.

credits.csv: Includes detailed information on the cast and crew associated with each title, such as names of actors, writers, and directors.

Together, these datasets enable a comprehensive exploration of the Amazon prime content ecosystem.

Tools and Libraries Used
The analysis was carried out using the following Python libraries:

Pandas: For loading, cleaning, and manipulating tabular data.

NumPy: For handling numerical operations and missing values.

Matplotlib: For creating basic plots and visualizations.

Seaborn: For advanced and visually appealing statistical visualizations.

These libraries allow for efficient data processing and visualization, making them ideal tools for exploratory data analysis.

Methodology
The project followed a structured EDA workflow:

Data Loading and Cleaning:

The CSV files were extracted from zip files and loaded into Pandas DataFrames.

Basic inspection using .head(), .info(), and .describe() helped identify data types, missing values, and overall structure.

Missing or inconsistent data was handled using NumPy and Pandas techniques.

Data Merging:

The two datasets were merged using a common key to allow combined analysis of title data with credits data.

Univariate Analysis:

The distribution of content types was analyzed (Movies vs TV Shows).

Frequency distributions of release years, genres, countries, and IMDb scores were explored.

Most common production countries and genres were identified.

Bivariate and Multivariate Analysis:

Relationships between runtime and IMDb scores were analyzed.

Score comparisons across different content types and genres were visualized.

Frequent directors and actors in Amazon prime content were identified and ranked.

Data Visualization:

Bar charts, histograms, box plots, and heatmaps were created using Seaborn and Matplotlib.

These visualizations provided intuitive insights into data trends, such as score distributions and content popularity over time.

Key Insights
The analysis revealed several interesting insights:

Movies form a majority of the content available on Amazon prime.

The most popular genres are Drama and Comedy.

The average IMDb rating typically falls between 6.0 and 7.0.

A few directors and actors appear frequently, indicating high productivity or popularity.

Conclusion
This project demonstrates the practical use of Python libraries for conducting an in-depth exploratory analysis on real-world datasets. By combining and analyzing title metadata and credit information, valuable insights into Amazon prime’s content trends and patterns were discovered. ## 3. ***Data Wrangling***

### Data Wrangling Code

````python
import pandas as pd
import ast

# Load data
titles = pd.read_csv("C:\\Users\\prave\\Desktop\\titles.csv")
credits = pd.read_csv("C:\\Users\\prave\\Desktop\\credits.csv")

# Fill missing values
titles['age_certification'] = titles['age_certification'].fillna('UNKNOWN')
titles['seasons'] = titles['seasons'].fillna(0).astype(int)
titles['imdb_score'] = titles['imdb_score'].fillna(titles['imdb_score'].mean())
titles['tmdb_score'] = titles['tmdb_score'].fillna(titles['tmdb_score'].mean())

# Convert 'genres' and 'production_countries' from string to list
titles['genres'] = titles['genres'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])
titles['production_countries'] = titles['production_countries'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else [])

# Merge titles and credits
merged = pd.merge(titles, credits, on='id', how='left')

# Show result
print("Titles shape:", titles.shape)
print("Credits shape:", credits.shape)
print("Merged shape:", merged.shape)
````
